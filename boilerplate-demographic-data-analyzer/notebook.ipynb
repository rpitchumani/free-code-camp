```python
import pandas as pd
```

```python
# Read data from file
df = pd.read_csv("adult.data.csv")
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = df["race"].value_counts()

race_count
```

```python
df
```

```python
# What is the average age of men?
average_age_men = df[df["sex"] == "Male"]["age"].mean(skipna=True)
average_age_men
```

```python
# What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = df["education"].value_counts(normalize=True).loc["Bachelors"]
percentage_bachelors
```

```python
print(df["education"].unique())
print(df["salary"].unique())
```

```python
 # What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
  # What percentage of people without advanced education make more than 50K?

list_higher_eduction = ["Bachelors", "Masters", "Doctorate"]

# with and without `Bachelors`, `Masters`, or `Doctorate`
higher_education = df[df["education"].isin(list_higher_eduction)]
lower_education = df[~df["education"].isin(list_higher_eduction)]

higher_education                     
```

```python
lower_education
```

```python
higher_education_rich = higher_education["salary"].value_counts(normalize=True).loc[">50K"]*100
higher_education_rich
```

```python
lower_education_rich = lower_education["salary"].value_counts(normalize=True).loc[">50K"]*100
lower_education_rich
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = df["hours-per-week"].min()
min_work_hours
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = df["hours-per-week"].value_counts().loc[min_work_hours]

print(num_min_workers)

rich_percentage = df[df["hours-per-week"]==min_work_hours]["salary"].value_counts(normalize=True).loc[">50K"]*100
print(rich_percentage)
```

```python
df_high_earnings_by_country = df.loc[df["salary"]==">50K", "native-country"].value_counts()
df_all_countries = df["native-country"].value_counts()

df_high_earnings_country_percent = (df_high_earnings_by_country/df_all_countries).fillna(0).sort_values(ascending=False)
df_high_earnings_country_percent
```

```python
# What country has the highest percentage of people that earn >50K?
highest_earning_country = df_high_earnings_country_percent.index[0]
highest_earning_country
```

```python
highest_earning_country_percentage = df[df["native-country"]==highest_earning_country]["salary"].value_counts(normalize=True).loc[">50K"] * 100
highest_earning_country_percentage
```

```python
# Identify the most popular occupation for those who earn >50K in India.
top_IN_occupation = df[((df["native-country"]=="India") & (df["salary"]==">50K"))]["occupation"].value_counts().sort_values(ascending=False).index[0]
top_IN_occupation
```
